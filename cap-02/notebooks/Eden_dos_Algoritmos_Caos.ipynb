# 🌪️ Sistema de Lorenz

Criado por Edward Lorenz ao estudar convecção atmosférica.

Equações:
dx/dt = σ (y − x)
dy/dt = x (ρ − z) − y
dz/dt = xy − βz

Pequenas mudanças nas condições iniciais geram trajetórias completamente diferentes.
Esse fenômeno ficou conhecido como Efeito Borboleta.


```python
import numpy as np
import matplotlib.pyplot as plt

sigma = 10
rho = 28
beta = 8/3

dt = 0.01
steps = 10000

xs = np.zeros(steps)
ys = np.zeros(steps)
zs = np.zeros(steps)

xs[0], ys[0], zs[0] = (0., 1., 1.05)

for i in range(steps - 1):
    xs[i+1] = xs[i] + sigma*(ys[i] - xs[i]) * dt
    ys[i+1] = ys[i] + (xs[i]*(rho - zs[i]) - ys[i]) * dt
    zs[i+1] = zs[i] + (xs[i]*ys[i] - beta*zs[i]) * dt

fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.plot(xs, ys, zs)
plt.title('Atrator de Lorenz')
plt.show()
```

# 🌀 Sistema de Rössler

Proposto por Otto Rössler (1976).

Equações:
dx/dt = -y - z
dy/dt = x + a y
dz/dt = b + z(x - c)

Sistema mais simples que Lorenz, mas também capaz de gerar caos.


```python
a = 0.2
b = 0.2
c = 5.7

dt = 0.01
steps = 10000

xs = np.zeros(steps)
ys = np.zeros(steps)
zs = np.zeros(steps)

xs[0], ys[0], zs[0] = (0., 1., 1.)

for i in range(steps - 1):
    xs[i+1] = xs[i] + (-ys[i] - zs[i]) * dt
    ys[i+1] = ys[i] + (xs[i] + a*ys[i]) * dt
    zs[i+1] = zs[i] + (b + zs[i]*(xs[i] - c)) * dt

fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.plot(xs, ys, zs)
plt.title('Atrator de Rössler')
plt.show()
```

# 🌌 Atrator Estranho

Um atrator estranho é:

- Um conjunto para onde o sistema converge
- Não periódico
- Com estrutura fractal
- Sensível às condições iniciais

Formalmente, possui pelo menos um expoente de Lyapunov positivo.


# 🦋 Sensibilidade às Condições Iniciais

Duas trajetórias iniciadas muito próximas divergem exponencialmente:

d(t) ≈ d(0) e^(λt)

Se λ > 0 → comportamento caótico.


## 2.3 - BIFURCAÇÃO: CAMINHO EM DIREÇÃO AO CAOS

## 🌿 Caminho em direção ao caos (Bifurcação)

### 🔹 Ponto fixo (estável)
O sistema **“para”** em um valor final.

---

### 🔹 Ciclo de período 2
Em vez de parar em um único valor, ele alterna entre **dois valores**.

---

### 🔹 Período 4
Agora alterna entre **quatro valores**.

---

### 🔹 Período 8, 16, 32…
O comportamento vai **dobrando sucessivamente**.

---

### 🔹 Caos
Chega um momento em que não é mais possível dizer  
“ele repete em 2, 4, 8…”

O sistema torna-se:

- Aperiodico  
- Sensível às condições iniciais  
- Imprevisível no longo prazo  

---

> 🔥 A sequência **2 → 4 → 8 → 16 → ...** é conhecida como a  
> **“estrada para o caos”** (rota de duplicação de período).

### Exemplo - Logistica

```python
# Logistica
import matplotlib.pyplot as fig

r=1
a=0.1
x=[a]

for k in range(100):
        a = r*a*(1-a)
        x.append(a)  

fig.plot(x,'-k',linewidth=2)
fig.xlabel('Iteração', fontsize=16)
fig.ylabel('x[k]', fontsize=24)
fig.title('Logistica com r = 3.8', fontsize=16)
```
