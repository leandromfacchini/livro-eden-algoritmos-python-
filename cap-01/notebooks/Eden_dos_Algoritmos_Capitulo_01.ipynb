# 🌱 O que é um Algoritmo?

Um algoritmo é uma sequência finita e bem definida de passos para resolver um problema.

Características principais:
- Finito (termina)
- Determinístico (mesma entrada → mesma saída)
- Bem definido

Algoritmos existem muito antes dos computadores. Um exemplo clássico é o Algoritmo de Euclides para cálculo do Máximo Divisor Comum (MDC).


```python
# Algoritmo de Euclides - cálculo do MDC
def mdc(a, b):
    while b != 0:
        a, b = b, a % b
    return a

print(mdc(48, 18))
```

# 🏛️ A Metáfora do Éden

O autor utiliza a metáfora do 'Éden' para representar o universo dos algoritmos.

Assim como as leis da natureza governam o mundo físico,
os algoritmos governam o mundo digital.

Tudo que envolve automação, recomendação, busca ou decisão computacional depende de algoritmos.


# 🧠 Algoritmos e o Pensamento Humano

O capítulo propõe que o pensamento humano pode ser interpretado como execução de algoritmos mentais.

Entrada (input) → Processamento → Saída (output)

Isso prepara o terreno para temas futuros como inteligência artificial e aprendizado de máquina.


```python
# Exemplo simples de algoritmo do cotidiano
def decidir_roupa(temperatura):
    if temperatura < 18:
        return "Use um casaco."
    elif temperatura < 25:
        return "Vista uma camiseta."
    else:
        return "Use roupas leves."

print(decidir_roupa(22))
```

# 🐍 Por que Python?

Python é escolhida por sua simplicidade e clareza.

Ela permite focar na lógica do algoritmo, sem distrações de sintaxe complexa.

Exemplo simples:


```python
def soma(a, b):
    return a + b

print(soma(2, 3))
```
